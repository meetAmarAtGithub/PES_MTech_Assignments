The dataset has a total of <b> 768 rows </b> each referring to patients characterized by <b> 9 numeric variables </b>. The data refers to females of ages from 21 to 81 years. The nine attributes can be further classified as 

A)	Six attributes describe the result of physical / verbal examination<br/>
    &emsp;
    1)	Number of times pregnant<br/>
    &emsp;
    2)	Diastolic blood pressure (mm Hg)<br/>
    &emsp;
    3)	Triceps skin fold thickness (mm)<br/>
    &emsp;
    4)	Body mass index (weight in kg/(height in m)2)<br/>
    &emsp;
    5)	Diabetes pedigree function - a function which scores probability of getting diabetes based on family history<br/>
    &emsp;
    6)	Age (years)<br/><br/>
    
   B)	Two attributes are result of chemical examination<br/>
   &emsp;
    7)	Plasma glucose concentration over 2 hours in an oral glucose tolerance test<br/>
    &emsp;
    8)	2-Hour serum insulin (mu U/ml)<br/><br/>
    
   C)	Resultant Target Variable<br/>
   &emsp;
    9)	Class variable -0 (No) if non-diabetic, 1 (Yes) if diabetic<br/><br/>




<div style="font-family: 'Roboto', serif; font-size: 20px">Objective:</div> of this data analysis is to find the significant factors in the first 8 variables which influence the 9th variable that is if (Resultant)(diabetes=yes (1) diabetes = no (0)) through data analysis and visualization techniques using Python<br/><br/>

# Finding the optimial values to replace in skin thickness and insulin levels to replace zero values 


```python
# create a temporary dataframe removing zero values to avoid skew in mean and median
df2 = df1[df1['2-Hour serum insulin (mu U/ml)'] > 0]
print(df2.isin([0]).sum())
```

```python
# find the grouped mean and median of insulin level
grouped_df =df2.groupby('Class variable')
smean = round(grouped_df['2-Hour serum insulin (mu U/ml)'].mean(),2)
smedian =round(grouped_df['2-Hour serum insulin (mu U/ml)'].median(),2)
```

```python
sns.boxplot(y='2-Hour serum insulin (mu U/ml)', x='Class variable', 
                 data=df2, 
                 palette="colorblind",
                 hue='Class variable',showmeans=True,
           meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"10"})
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.text(0,smean.loc[0],str((smean.loc[0])))
plt.text(1*.75,smean.loc[1],str((smean.loc[1])))
plt.text(0,smedian.loc[0],str((smedian.loc[0])))
plt.text(1*.75,smedian.loc[1],str((smedian.loc[1])))
```

```python
# replace skin thickness zeros
df1['Triceps skin fold thickness (mm)'] = np.where(((df1['Triceps skin fold thickness (mm)']==0)&(df1['Class variable']==0)), 27.21, df1['Triceps skin fold thickness (mm)'])
df1['Triceps skin fold thickness (mm)'] = np.where(((df1['Triceps skin fold thickness (mm)']==0)&(df1['Class variable']==1)), 32.96, df1['Triceps skin fold thickness (mm)'])
```

```python
# replace insulin level zero values
df1['2-Hour serum insulin (mu U/ml)'] = np.where(((df1['2-Hour serum insulin (mu U/ml)']==0)&(df1['Class variable']==0)), 102.5, df1['2-Hour serum insulin (mu U/ml)'])
df1['2-Hour serum insulin (mu U/ml)'] = np.where(((df1['2-Hour serum insulin (mu U/ml)']==0)&(df1['Class variable']==1)), 169.5, df1['2-Hour serum insulin (mu U/ml)'])
```

```python
# drop zero values in plasma glucose
df1 =df1[df1['Plasma glucose concentration a 2 hours in an oral glucose tolerance test'] > 0]
## df1.info()
```

```python
# drop zero values in blood pressure
df1 =df1[df1['Diastolic blood pressure (mm Hg)'] > 0]
##df1.info()
```

```python
# drop zero values in BMI
df1 =df1[df1['Body mass index (weight in kg/(height in m)^2)'] > 0]
##df1.info()
```

```python
#This is to find how many column wise zero(0) rows are there!
print(df1.isin([0]).sum())
```
